# DATA CLEANING
---

## Step 1 : Make a Copy of the Raw Data

```python
df_raw = df.copy()   # keep original untouched
df_clean = df.copy() # all cleaning done on this copy

print('Raw shape   :', df_raw.shape)
print('Clean shape :', df_clean.shape)
```

## Step 2 : Fix Data Types — Convert Date Columns (object → datetime)

```python
date_cols = ['orderDate', 'requiredDate', 'shippedDate']

for col in date_cols:
    df_clean[col] = pd.to_datetime(df_clean[col], errors='coerce')

print('Date column dtypes after conversion:')
print(df_clean[date_cols].dtypes)
```

## Step 3 : Fix Data Types — discontinued Column (int → bool)

```python
# Before
print('Before:', df_clean['discontinued'].dtype, '| Values:', df_clean['discontinued'].unique())

df_clean['discontinued'] = df_clean['discontinued'].astype(bool)

# After
print('After :', df_clean['discontinued'].dtype, '| Values:', df_clean['discontinued'].unique())
```

## Step 4 : Fix Data Types — reports_to Column (float → Int64, nullable integer)

```python
# reports_to is float because of NaN values
# Use pandas nullable integer type Int64 (capital I) to keep NaN support
print('Before:', df_clean['reports_to'].dtype)
print(df_clean['reports_to'].value_counts(dropna=False))

df_clean['reports_to'] = df_clean['reports_to'].astype('Int64')

print('After :', df_clean['reports_to'].dtype)
```

## Step 5 : Fix Floating Point Precision in line_total

```python
# Example of floating point issue: 1261.3999999999999 should be 1261.40
print('Before rounding (sample):')
print(df_clean[df_clean['line_total'] != round(df_clean['line_total'], 2)]['line_total'].head(5).values)

df_clean['line_total'] = df_clean['line_total'].round(2)

print('After rounding (sample):')
print(df_clean['line_total'].head(5).values)
```

## Step 6 : Remove Redundant Column — product_unitPrice

```python
# unitPrice (from order_details) = actual price at time of order
# product_unitPrice (from products) = current catalogue price
# We keep unitPrice (transactional) and drop product_unitPrice (catalogue)

print('Columns before:', df_clean.shape[1])
df_clean.drop(columns=['product_unitPrice'], inplace=True)
print('Columns after :', df_clean.shape[1])
print('Dropped: product_unitPrice')
```

## Step 7 : Handle Missing Values — shippedDate and ship_delay_days

> These 73 rows are orders that have NOT yet been shipped (no ship date = valid business reason).  
> We DO NOT drop them — instead we flag them with a new column `is_shipped`.

```python
# Flag unshipped orders
df_clean['is_shipped'] = df_clean['shippedDate'].notna()

print('Shipped orders     :', df_clean['is_shipped'].sum())
print('Not shipped orders :', (~df_clean['is_shipped']).sum())

# ship_delay_days is NaN where order is not shipped — correct, leave as-is
print('Missing shippedDate    :', df_clean['shippedDate'].isnull().sum())
print('Missing ship_delay_days:', df_clean['ship_delay_days'].isnull().sum())
```

## Step 8 : Handle Missing Values — reports_to

> `reports_to` is null for the top-level manager (no one to report to).  
> This is intentional — we fill with 0 to indicate no manager.

```python
print('Missing reports_to before:', df_clean['reports_to'].isnull().sum())

df_clean['reports_to'] = df_clean['reports_to'].fillna(0).astype('Int64')

print('Missing reports_to after :', df_clean['reports_to'].isnull().sum())
print('Value counts:')
print(df_clean['reports_to'].value_counts())
```

## Step 9 : Remove Duplicate Rows (if any)

```python
before = df_clean.shape[0]
df_clean.drop_duplicates(inplace=True)
after = df_clean.shape[0]

print(f'Rows before : {before}')
print(f'Rows after  : {after}')
print(f'Duplicates removed: {before - after}')
```

## Step 10 : Standardize Text Columns — Strip Whitespace

```python
text_cols = df_clean.select_dtypes(include='object').columns.tolist()

for col in text_cols:
    df_clean[col] = df_clean[col].str.strip()

print(f'Stripped whitespace from {len(text_cols)} text columns:')
print(text_cols)
```

## Step 11 : Validate — Quantity and Price Must Be Positive

```python
invalid_qty   = df_clean[df_clean['quantity'] <= 0]
invalid_price = df_clean[df_clean['unitPrice'] <= 0]
invalid_total = df_clean[df_clean['line_total'] < 0]

print(f'Rows with quantity <= 0  : {len(invalid_qty)}')
print(f'Rows with unitPrice <= 0 : {len(invalid_price)}')
print(f'Rows with line_total < 0 : {len(invalid_total)}')
```

## Step 12 : Validate — Discount Must Be Between 0 and 1

```python
invalid_discount = df_clean[(df_clean['discount'] < 0) | (df_clean['discount'] > 1)]

print(f'Rows with invalid discount (outside 0-1): {len(invalid_discount)}')
print('Discount value range:')
print(df_clean['discount'].describe())
```

## Step 13 : Validate — orderDate Must Be Before shippedDate

```python
shipped_rows = df_clean[df_clean['shippedDate'].notna()].copy()
invalid_dates = shipped_rows[shipped_rows['shippedDate'] < shipped_rows['orderDate']]

print(f'Orders where shippedDate < orderDate (invalid): {len(invalid_dates)}')
if len(invalid_dates) == 0:
    print('All shipping dates are valid.')
```

## Step 14 : Final Null Check After Cleaning

```python
remaining_nulls = df_clean.isnull().sum()
remaining_nulls = remaining_nulls[remaining_nulls > 0]

if remaining_nulls.empty:
    print('No missing values remaining.')
else:
    print('Remaining missing values:')
    print(remaining_nulls)
```

## Step 15 : Cleaned Dataset — Final Info

```python
print('=== CLEANED DATASET SUMMARY ===')
print(f'Shape        : {df_clean.shape}')
print(f'Total Orders : {df_clean["orderID"].nunique():,}')
print(f'Date Range   : {df_clean["orderDate"].min().date()} to {df_clean["orderDate"].max().date()}')
print(f'Total Revenue: ${df_clean["line_total"].sum():,.2f}')
print()
df_clean.info()
```
